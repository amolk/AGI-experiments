```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import norm

sns.set(style="white", palette="muted", color_codes=True)

%matplotlib inline

u = 16
sigma_u = 1
v_p = 3
sigma_p = 1
v_step = 0.01 # step size
v_range = np.arange(0.01, 5, v_step) # range of sensory input

prior = norm.pdf(v_range, v_p, sigma_p)
likelihood = norm.pdf(u, v_range * v_range, sigma_u)
numerator = np.multiply(prior, likelihood)

denominator = sum(numerator * v_step)

posterior = numerator / denominator

plt.figure(figsize=(7.5, 2.5))
plt.plot(v_range, prior)
plt.ylabel('prior')
sns.despine()

plt.figure(figsize=(7.5, 2.5))
plt.plot(v_range, likelihood)
plt.ylabel('likelihood')
sns.despine()


plt.figure(figsize=(7.5,2.5))
plt.plot(v_range, posterior)
plt.ylabel(r' $p(v | u)$')
sns.despine()
```
