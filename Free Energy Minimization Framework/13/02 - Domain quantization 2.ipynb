```python
import torch
from torch import nn
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.stats import norm
import pandas as pd

def normal_distribution(n, mean, var=0.05):
    x = norm.pdf(np.arange(0, 1, 1.0 / n), mean, var)
    x = x / np.sum(x)
    #return torch.tensor(x).float()
    return x

class DomainQuantization:
    def __init__(self, bin_count=10, init_low_bound=0.0, init_high_bound=1.0):
        assert bin_count > 1

        self.bin_count = bin_count
        self.bins = None # np.linspace(init_low_bound, init_high_bound, bin_count)

    def qcut_bins(self, values):
        assert len(values.shape) == 1, "Need 1D numpy array of values"
        _, bins = pd.qcut(values, q=self.bin_count-1, retbins=True)
        return bins

    def forward(self, values):
        digits = np.digitize(values, self.bins)
        output = digits.astype(float)
        for index, digit in enumerate(digits):
            if digit > 0 and digit < len(self.bins):
                val_left = self.bins[digit-1]
                val_right = self.bins[digit]
                output[index] += (values[index] - val_left) / (val_right - val_left)

        output -= 1
        output.clip(0, len(self.bins) - 1, out=output)
        return output
    
    def update_bins(self, values):
        alpha = 0.8
        bins = self.qcut_bins(values)
        if self.bins is None:
            self.bins = bins
        else:
            self.bins = self.bins * alpha + bins * (1 - alpha)

        
def create_test_pattern(pattern_length=200):
    a = torch.tensor((np.sin(np.arange(pattern_length) * 0.20) + np.cos(np.arange(pattern_length) * 0.10) + np.random.sample(int(pattern_length)) * 0.0) / 2.5).float()

    fig, ax1 = plt.subplots(figsize=(15, 5))
    plt.title('Pattern')
    ax1.plot(a.numpy())
    plt.show()

    return a


def create_pattern(pattern_length=200):
    p = torch.tensor((np.sin(np.arange(pattern_length) * 0.20) + np.random.sample(int(pattern_length)) * 0.0)).float()

    fig, ax1 = plt.subplots(figsize=(15, 5))
    plt.title('Test Pattern')
    ax1.plot(p.numpy())
    plt.show()

    return p


pattern_floats = create_pattern()
dq = DomainQuantization(bin_count=10)

dq.update_bins(pattern_floats)
fig, ax1 = plt.subplots(figsize=(15, 5))
ax1.plot(dq.forward(pattern_floats))
plt.show()
print(dq.bins)

test_pattern_floats = create_test_pattern()

for _ in range(30):
    fig, ax1 = plt.subplots(figsize=(15, 5))
    ax1.plot(dq.forward(test_pattern_floats))
    plt.show()
    
    #dq.update_bins(test_pattern_floats)
    print(dq.bins)

    

# fig, ax1 = plt.subplots()
# # plt.title('Values use full range of prior => full resolution')
# ax1.plot(pattern_floats)
# ax2 = ax1.twinx()
# ax2.plot(scaled_posterior(values, quantization), 'r')
# plt.show()

# print(dq.find_range_markers(pattern_floats))
# print(dq.update(pattern_floats))
```

Domain quantization converts non-linear functions to a piecewise linear functions, while minimizing error for a given number of bins.
